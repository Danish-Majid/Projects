<h3> 3. Ageing Population </h3>
<p> Ageing population happens when there are declining birth rates combined with an increase in life expectancy. Declining birth rates happens during the <a href="https://mothership.sg/2018/05/singapore-stop-at-two-children-campaign-successful/"> “Two-is-Enough” population control campaign of the early 1970s</a>. However, even after 40 years and much higher birth rates, there is still an increase in the ageing population which is projected to be 25% of the total population in 2030.<p> You might ask, why is there still an increase after all these years? The answer is due to the increase in life expectancy.</p>

<h4>3.1 Health Adjusted Life Expectancy At Birth (HALE) Dataset</h4>
<p>The first dataset I will be using is from data.gov.sg database which is titled <a href="https://data.gov.sg/dataset/health-adjusted-life-expectancy?view_id=1f3acfb6-b53f-4974-8088-ffc3b069cade&resource_id=07768e9d-222d-466e-a786-e2f90b33c057"> Health Adjusted Life Expectancy at Birth</a> which I have loaded in as a Pandas Dataframe called <code>hale</code>. I chose this dataset as I want to show the current life expectancy of Singaporeans and how it will impact the ageing population.</p>

<p>Health Adjusted Life Expentancy at Birth (HALE) is taken from <a href="https://www.moh.gov.sg/docs/librariesprovider5/default-document-library/gbd_2017_singapore_reportce6bb0b3ad1a49c19ee6ebadc1273b18.pdf"> Global Burden of Disease (GBD) 2017 Study</a> which reports its findings from 1990 - 2017. It is interpreted as the number of years from birth in which an individual can  expect to live healthy before getting disablilities and chronic diseases.</p>

<p>Descriptions of the columns:</p>
<ul>
    <li> <code>sex</code> : Gender </li>
    <li> <code>hale_years</code> : Number of years a person is expected to live healthy without any disabilities or illnesses </li>
</ul>
<p> I used <code>.describe()</code> to describe the dataset better. For the HALE years of Singaporeans, it has a mean of 71 HALE years, have a minimum of 65.5 HALE years and maximum of 75.8 HALE years.</p>

```python
hale.info()
print('\n', 70*'*')
display(hale.describe())
```

<h4>3.2 HALE Line Graph (1)</h4>
<p> The graph below shows the HALE years after birth from 1990 to 2017. As you can see, the number of years a person is expected to be healthy has been steadly increasing from around 65.6 to 72.6 for males and 68.5 to 75.8 for females. This is mainly due to the improved healthcare situation in Singapore. The evidence for this is when I will be talking about mortality rate dataset.</p>
<p>For the graph, I used a line plot with the colours <code>'dodgerblue'</code> and <code>'hotpink'</code> to make it very distinguishable between the Male and Female data. I also used a <code>year</code> list where it has an increment of 3 years and used <code>plt.xticks()</code> to assign the years accordingly.</p>

```python
#Creating group for column 'sex' and finding HALE years of the 2 genders
gender = hale.groupby('sex')
male = gender.get_group('Male')['hale_years']
female = gender.get_group('Female')['hale_years']

#Creating list for years with intervals of 3
year = [1990]
for x in np.arange(0,9):
    year = np.append(year, year[x]+3)

plt.figure(figsize=(15, 5))

#Plotting Line Graphs
male_plot, = plt.plot(male.index, male, color = 'dodgerblue')
female_plot, = plt.plot(female.index, female, color = 'hotpink')

#Setting labels
plt.xlabel('Year')
plt.ylabel('HALE Years')
#Setting title
plt.title('Health Adjusted Life Expectancy of Singaporeans from 1990 - 2017')
#Setting legend
plt.legend([male_plot, female_plot],['Male', 'Female'])
#Setting x ticks
plt.xticks( year)

plt.show()
```

<h4>3.3 HALE Bar Graph (2)</h4>
<p>Intially, I wanted to create a graph like a Population Pyramid graph which have 2 sides where one side will be Males and the other is Female. I made it using <code>plt.barh()</code> where it will plot a bar graph horizontally. To create 2 sides, i put a <code>-</code> infront of <code>female</code>. Due to the small gap in between 2015 and 2017, I decided a line graph is better to interpret the data clearly.</p>

```python
#Create group for the two genders
male = gender.get_group('Male')['hale_years']
female = gender.get_group('Female')['hale_years']

plt.figure(figsize=(15, 5))

#plt.barh creates a bar graph horizontally
plt.barh(male.index, male, height = 1, color = 'dodgerblue')
plt.barh(female.index, -female, height = 1, color = 'hotpink')

#Make x-axis to be positive values only
plt.xticks(np.arange(-80,80,20),[80,60,40,20,0,20,40,60,80])

#Setting Legend
plt.legend([male_plot, female_plot],['Male', 'Female'])
#Setting Title
plt.title('Health Adjusted Life Expectancy of Singaporeans from 1990 - 2017')
#Setting Labels
plt.ylabel('Year')
plt.xlabel('HALE Years')
plt.show()
```

<h4>4.2 Healthcare Expenditure Line Graphs</h4>
<p> From this graph, we can observe the trends of the different categories of spending by the government. As you can see, the government health expenditure and the operating expenditure have been steadily increasing. However, the development expenditure have been influctuating with recently a drop of about 200 Million. The evidence of this is that when the development expenditure decreased, the percentage GDP remained the same, meaning that funds have increased significantly for the other categories.</p>

<p>For this graph, I used a 2x2 subplot which I created using <code>plt.subplots(2,2)</code>. I also used <code>sharex = True</code> to share the x-axis of bottom graph with the top graph since the years are the same. <code>fig, a</code> is a shorter version of <code>plt.figure() and a = plt.subplots</code>. This to make it easier for me to plot the graphs by using <code>a</code> instead of <code>plt</code>. </p>

<p>For annotating the graph to iterate the point I am making, I used <code>.annotate()</code>. It is my first time using this so I had trouble creating the arrows. I found a website, <a href="https://www.programcreek.com/python/example/102284/matplotlib.pyplot.annotate"> matplotlib.pyplot.annotate() examples</a>, which shows a few examples on how to use it. I used the arrow style which is shown in the website.</p>

```python
#Setting subplots, size of graph and sharing of x-axis
fig, a = plt.subplots(2,2, figsize=(15, 5), sharex=True)

#Plotting Development Expenditure graph
a[0,0].plot(health_ex.index,health_ex['development_expenditure'], 'tab:blue')
a[0,0].set_title('Development Expenditure')
a[0,0].set_ylabel('In Millions SGD')
a[0,0].annotate('Drop', xy=(2016, 1600),xytext=(2016, 1200),
                arrowprops={'arrowstyle': '->', 'shrinkA': 1, 'shrinkB': 1, 'lw':1.0,'mutation_scale': 15., 'color': 'black'})

#Plotting Operating Expenditure graph
a[0,1].plot(health_ex.index,health_ex['operating_expenditure'], 'tab:cyan')
a[0,1].set_title('Operating Expenditure')
a[0,1].set_ylabel('In Millions SGD')

#Plotting Government Health Expenditure graph
a[1,0].plot(health_ex.index,health_ex['government_health_expenditure'], 'tab:red')
a[1,0].set_title('Government Health Expenditure')
a[1,0].set_ylabel('In Millions SGD')

#Plotting Percentage of GDP graph
a[1,1].plot(health_ex.index,health_ex['percentage_gdp'], 'tab:orange')
a[1,1].set_title('Percentage of GDP')
a[1,1].set_ylabel('In Percentage')
a[1,1].annotate('No Drop', xy=(2016, 2.1),xytext=(2016, 1.5),
                arrowprops={'arrowstyle': '->', 'shrinkA': 1, 'shrinkB': 1, 'lw':1.0,'mutation_scale': 15., 'color': 'black'})

plt.show()
```

<h4>4.3 Development and Operating Expenditure Bar Graph</h4>
<p> To answer the question that I posed, I needed data from the operating and development expenditure. For operating expenditure, it shows that it has increased for 11 years straight without it ever decreasing. The expenditure kept on increasing due to the government expanding clinical services and capacity, improving quality through the training and retention of
healthcare professionals and funding for programmes from HPB. It is also due to the increasing demand for health and aged care services. This shows that the government is taken note of the ageing population and increased funding to meet the demands of the population.</p>

<p>For the development expenditure graph, it is not as smooth as the operating expenditure. This data is mainly to show the amount of money the government have spent to fund major projects. Major projects include the construction of new hospitals, new IT infrastructures and laboratories.</p>

    <li>For the drop in 2017, it is mainly due to lower cashflows required for projects such as the National Centre for Infectious Diseases and the Integrated Development of Sengkang General and Community Hospital which are nearing completion.
</ul>
<p>This means that between 2010 to 2017, the government have funded more constructions of hospitals and facilities. This could be linked back to the question where there are preparing more facilities and more beds for the future ageing population.</p>

<h5>4.4.1 Preprocessing Part 1</h5>
<p> While exploring the DataFrame <code>facility</code> I found out that it will be better to show the total facilities and beds instead of each data from the different types of facilities.</p>
<p>For the Dataset, the columns are seperated at an interval of 5 for each of the years. Therefore, I created a for loop and a counter <code>i</code>. I used an if statement where when <code>i != 5</code> it will add the number of facilities and beds into <code>sum_fac</code> and <code>sum_bed</code> respectively. The counter will then increase by 1. Once <code>i = 5</code>, the counter values will be stored in a list and <code>i = 0</code>. It will then loop until the end of the DataFrame</p>
<p> After I have the total of the number of facilities and beds, I created a dictionary to store the list to the column name and used <code>pd.DataFrame()</code> to create the DataFrame <code>fac_total</code>.</p>

```python
#Original DataFrame
display(facility.head())
print('\n', 70*'*')

#Convert Dataframe to numpy array
fac_new = facility.to_numpy()

#Create lists
total_bed = []
total_fac = []
years = np.arange(2009,2020)

#Counters
sum_bed = 0
sum_fac = 0
i=0

#For loop to add number of facilities and beds for each year
for x in np.arange(0,55):
    if i != 5:
        sum_bed += fac_new[x,3]
        sum_fac += fac_new[x,2]
        i += 1
        if i == 5:
            total_bed = np.append(total_bed, sum_bed)
            total_fac = np.append(total_fac, sum_fac)
            sum_bed = 0
            sum_fac = 0
            i = 0
            
#Create dictionary
data = {'year': years,'total_bed' : total_bed, 'total_facility' : total_fac}

#Create DataFrame from Dictionary
fac_total = pd.DataFrame(data, columns = ['year','total_facility', 'total_bed'])
display(fac_total.tail())
```

<p>For this graph, I made a scatter plot using the x-axis to be the number of facilities and y-axis to be number of beds. I used a scatter plot to show the corelation between the two axis. This code is the hardest to make user friendly. I tried using For loops to create the graph but I lacked the control to plot the legend. I also tried defining functions but I ended up confusing myself. So, I made it simple by using a function <code>find</code> to take in the <code>group_var</code>, which is the variable containing the <code>.groupby()</code> function, and <code>num</code>, which is the index of the list <code>types</code>. I used a list of the types of facilities as I felt that it is very tedious to write out the full names of the facilities. To make it simpler, I could just call out the index.</p>

<p>I used <code>plt.scatter()</code> to create the scatter plot. I passed in <code>'no_of_facilities' and 'no_beds'</code> as the x and y axis respectively. I used <code>label</code> to create the name to be appear in the legend. I also used <code>alpha=0.6, s = 200</code>. <code>alpha</code> is to adjust the transparency level of the dots as I found out that some of the dots are overlapping each other and some dots could not be seen if I did not adjust the transparency levels. <code>s</code> is to control the size of the dots as the default size is quite small.</p>

```python
#Create lists
types = ['Acute', 'Psychiatric Hospitals', 'Community Hospitals', 'Nursing Homes', 'Inpatient Hospices']

#Create function to find different types of facilities data
find = lambda group_var, num: group_var.get_group(types[num])

#Setting size of graph
plt.figure(figsize=(15,5))

#Group the types of facilities
fac = facility.groupby('facility_type_a')

#Group the different types of facilities into a variable
acute = find(fac, 0)
psy = find(fac, 1)
com = find(fac, 2)
nur = find(fac, 3)
inp = find(fac, 4)

#Plotting scatter graph
acute_plot = plt.scatter(acute['no_of_facilities'],acute['no_beds'], label = types[0],  alpha=0.6, s = 200)
psy_plot = plt.scatter(psy['no_of_facilities'],psy['no_beds'], label = types[1],  alpha=0.6, s = 200)
com_plot = plt.scatter(com['no_of_facilities'],com['no_beds'], label = types[2],  alpha=0.6, s = 200)
nur_plot = plt.scatter(nur['no_of_facilities'],nur['no_beds'], label = types[3],  alpha=0.6, s = 200)
inp_plot = plt.scatter(inp['no_of_facilities'],inp['no_beds'], label = types[4],  alpha=0.6, s = 200)

#Titles, labels and legend
plt.title('Relationship between Number of Facilities and Number of Beds from 2009 to 2019')
plt.xlabel('Number of Facilities')
plt.ylabel('Number of Beds')
plt.legend(handles = [acute_plot,psy_plot,com_plot,nur_plot,inp_plot])

plt.show()
```
